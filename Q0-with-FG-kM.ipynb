```python
#MCMC data
k_index=0
data_mcmc = (k[k_index:], PCA_Q0[k_index:], sigmaQ[k_index:])

data_mcmc[0]
```

```python
def lnlike(theta, x, y, yerr):    
    return -0.5 * np.sum(((y - model_ph(theta, x))/yerr) ** 2)
```

```python
def lnprior(theta):
    
    THI,b1,b2,c1,kM = theta
    
    if 0. < THI < 0.5 and 0. < b1 < 4. and 0.< kM < 1.:
        return 0.0
    #if -10. < b1 < 10.:
    #    return 0.0
    #if -10. < b2 < 10.:
    #    return 0.0
    #if -100. < PN < 100.:
    #    return 0.0
    
    return -np.inf
```

```python
def lnprob(theta, x, y, yerr):
    lp = lnprior(theta)
    if not np.isfinite(lp):
        return -np.inf
    return lp + lnlike(theta, x, y, yerr)
```

```python
#kM = 0.07

THI_fid = 0.0737 #mK [confirmed by Steve]
bHI_fid = 1.1 #[confirmed by Steve]


#model from https://arxiv.org/pdf/1910.07097.pdf
def model_ph(theta,k):
    
    THI,b1,b2,c1,kM = theta
    res = THI**2 * (b1 + b2*k**2)**2 * Pk_halofit_z0p4(k) + P_noise*(c1-c1*(kM/k)**2)
    
    return res
```
